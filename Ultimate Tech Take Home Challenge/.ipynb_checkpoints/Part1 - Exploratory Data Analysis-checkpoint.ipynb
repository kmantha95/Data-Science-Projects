The goal here is to aggregate the login counts based on 15-minute time intervals, and visualize and describe the resulting time series of login counts in ways that best characterize the underlying patterns of the demand. This will be followed by reporting important features of the demand, such as daily cycles. Data quality issues should be reported as well.

```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
%matplotlib inline
import datetime
```

```python
login_times = pd.read_json('logins.json')
```

```python
login_times.head()
```

```python
login_times.info()
```

It appears that `login_time` is the only column, that too of type 'datetime' so we should be good to perform EDA within the context of time intervals on this dataset.

We can now aggregate the login times based on 15min intervals.

```python
login_times.set_index('login_time', inplace=True) #set dataset index to the login_time column
login_times['count'] = 1 #assign count to each timestamp
login_times.sort_index(inplace=True) #sort dataset
login_times.head()
```

```python
login_times_interval = login_times.resample('15min').sum() #count up each 15min interval
login_times_interval.head()
```

```python
login_times_interval.info()
```

There are 9788 entries going from Jan1 1970 until April13 1970. We can break the data down by full cycle, months, weeks, and days, to be able to analyze the login trends at both the macro- and micro- scopic levels.

```python
login_times_interval.plot()
plt.show()
```

The graph does not discern much at this level. However, it does show that demand fluctuates with some type of consistency. We can hypothesize that this is influenced by increased logins on weekends. We can zoom in on individual portions of the graph to better analyze the trends.

```python
login_times_interval.rolling(window=128).mean().plot()
plt.show()
```

There appears to be a weekly cycle with login times. Let's pick a month (March, let's say) to look at closer.

```python
login_times_interval.loc['1970-03'].plot()
plt.show()
```

```python
login_times_interval.loc['1970-03'].rolling(window=128).mean().plot()
plt.show()
```

There is clearly an overall rise in demand on weekends, however there are dips within those increases. We can hypothesize that the declines are due to a decrease in activity at night.

Let's get more microscopic and look at data for one week.

```python
datetime.datetime(1970, 3, 1).weekday()
```

This means the first day of the month was a Sunday (0 is Monday, 6 is Sunday)

```python
login_times_interval.loc['1970-03-01':'1970-03-07'].plot()
plt.show()
```

There does seem to be increase in login demand on the weekend compared to the weekdays, but it is still cyclical day by day. Let's take a look at a single weekend.

```python
login_times_interval.loc['1970-03-06':'1970-03-08'].plot()
plt.show()
```

It appears that demand is highest actually near midnight and the end of the day. There is a similar, but smaller, spike during noon/midday as well. Let's confirm this relationship for the weekdays as well as weekends.

```python
login_times_interval.loc['1970-03-03':'1970-03-04' ].plot()
plt.show()
```

The results hold true for the weekend as well as the weekdays.

There are 2 overarching conclusions here:
Firstly, login demand is greater on the weekends than it is during the week.
Secondly, demand appears highest around midnight on any given day, with a similar but smaller seen around noon, and this trend is seen on both weekdays and weekends.
