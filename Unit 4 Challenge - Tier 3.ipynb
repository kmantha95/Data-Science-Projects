**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
#make a function that gets the ratio of the prices of all boroughs from 2000 until 2020
def price_ratio(d):
    y2000 = float(d['Average_Price'][d['Year']==2000])
    y2020 = float(d['Average_Price'][d['Year']==2020])
    ratio = [y2020/y2000]
    return ratio
```

```python
price_ratio(df_grouped[df_grouped['London_Boroughs']=='Camden']) #testing function
```

```python
final_ratios = {} #make an empty dictionary to store the 2020:2000 ratios of prices for each borough

for b in df_grouped['London_Boroughs'].unique(): #iterate through each borough, and store with its ratio in final_ratios
    borough = df_grouped[df_grouped['London_Boroughs'] == b]
    final_ratios[b] = price_ratio(borough) 

print(final_ratios) #check to see that each borough's ratio stored properly
```

```python
df_ratios = pd.DataFrame(final_ratios) #make dataframe from final_ratios data to make it easier to see
df_ratios.head()
```

```python
df_ratios_T = df_ratios.T #transpose
df_ratios = df_ratios_T.reset_index() #reset index
df_ratios.head()
```

```python
df_ratios.rename(columns={'index':'Borough', 0:'2020'}, inplace=True) #cleaner names
top10 = df_ratios.sort_values(by='2020',ascending=False).head(10) #descending sort the top 10 ratios

print(top10)
```

```python
top10_plot = top10[['Borough','2020']].plot(kind='bar')
top10_plot.set_xticklabels(top10.Borough)
```
